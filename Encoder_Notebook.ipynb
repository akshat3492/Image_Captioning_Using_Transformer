```python
# Import necessary libraries and functions from custom scripts
import torch
import torchvision
from torch.utils.data import DataLoader, random_split
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import time

import warnings
warnings.filterwarnings("ignore")

from Encoder.Encoder_DataSetup import *
from Encoder.encoder_train import *
```

```python
# Define data transformations for training and validation

transform_train = transforms.Compose([
        # transforms.RandomCrop(32, padding=4),
        transforms.Resize((128, 128)),
        transforms.RandomHorizontalFlip(),
        transforms.RandomGrayscale(p=0.1),
        transforms.ToTensor(),
        transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010)),
    ])

test_train = transforms.Compose([
        transforms.Resize((128, 128)),
        transforms.ToTensor(),
        transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010))
    ])
```

```python
# Load datasets

train_dataset = CaptioningDataset("./data/flickr30k_images/Captions.csv", "./data/flickr30k_images/flickr30k_images/", transform_train)
val_dataset = CaptioningDataset("./data/flickr30k_images/Captions.csv", "./data/flickr30k_images/flickr30k_images/", test_train)
```

```python
# Split datasets into training and validation sets

total_ids = list(range(len(train_dataset)))
train_size = 0.8 #can easily customize
train_ids = random.sample(total_ids, int(train_size*len(total_ids)))
val_ids = list(set(total_ids).difference(train_ids))
```

```python
traindata = torch.utils.data.Subset(train_dataset, indices=train_ids)
valdata = torch.utils.data.Subset(val_dataset, indices=val_ids)
```

```python
# Set up data loaders

num_workers = 4
batch_size = 64

trainLoader = torch.utils.data.DataLoader(traindata, batch_size=batch_size, 
                                          num_workers=num_workers, drop_last=True, shuffle=True)
valLoader = torch.utils.data.DataLoader(valdata, batch_size=batch_size, 
                                          num_workers=num_workers, drop_last=True)
```

```python
import matplotlib.pyplot as plt

# Function to display images
def show_images(images, titles=None):
    """
    Displays images.
    """
    
    fig, axes = plt.subplots(1, len(images), figsize=(15, 5))
    if titles is None:
        titles = [''] * len(images)
    for img, title, ax in zip(images, titles, axes):

        # Clip the image to be within [0, 1] range
        img = torch.clamp(img, 0, 1)
        
        ax.imshow(img.permute(1, 2, 0).cpu().numpy())  # Convert to HWC format for plotting
        ax.set_title(title)
        ax.axis('off')
    plt.show()

batch_size = 3
sampleloader = torch.utils.data.DataLoader(traindata, batch_size=batch_size, 
                                          num_workers=num_workers, drop_last=True)

# Load a single batch from the DataLoader
data_iter = iter(sampleloader)
dataset = next(data_iter)

for i in range(batch_size):
    print(f"Rotation label: {dataset["Rotation Label"][i].item()}")
    show_images([dataset['Original Image'][i], dataset["Rotated Image"][i]], titles=["Original Image", "Rotated Image"])
    print(f"Captions: {[j[i] for j in dataset["Captions"]]}")
    print("\n")
```
